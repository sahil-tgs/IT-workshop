# Statistical Analysis

```python
#short summary of the data
df.info()
```

```python
#statistical summary of the data.
df.describe()
```

```python
# Calculate the survival status distribution
status_counts = df['status'].value_counts()

# Create subplots for each plot
fig, axis = plt.subplots(2, 2, figsize=(16, 10))

# Plot the age histogram
axis[0, 0].hist(df['age'], bins=10, color='red', alpha=0.7)
axis[0, 0].set_xlabel('age')
axis[0, 0].set_ylabel('frequency')
axis[0, 0].set_title('Age Histogram')

# Plot the year histogram
axis[0, 1].hist(df['year'], bins=10, color='green', alpha=0.7)
axis[0, 1].set_xlabel('year')
axis[0, 1].set_ylabel('frequency')
axis[0, 1].set_title('Year Histogram')

# Plot the nodes histogram
axis[1, 0].hist(df['nodes'], bins=10, color='blue', alpha=0.7)
axis[1, 0].set_xlabel('nodes')
axis[1, 0].set_ylabel('frequency')
axis[1, 0].set_title('Nodes Histogram')

# Plot the survival status pie chart
labels = ['Survived 5 years or longer', 'Died within 5 years']
axis[1, 1].pie(status_counts, labels=labels, startangle=90, autopct='%1.1f%%')
axis[1, 1].set_title('Survival Status Pie Chart')

# Show the plot
plt.tight_layout()
plt.show()
```

**Conclusion**

* There are a total of 306 entries of patient in the given data set

# Conclusive Analysis


**Node Vs Status**

```python
# Get the number of nodes for each status
status_1 = df[df['status'] == 1]['nodes']
status_2 = df[df['status'] == 2]['nodes']

# Count the frequency of each number of nodes in status 1
nodes_counts_1 = status_1.value_counts()
nodes_counts_2 = status_2.value_counts()

# Create a subplot for each pie chart
fig, (ax1, ax2, ax3) = plt.subplots(1, 3, figsize=(15, 5))

# Plot the third bar chart
ax1.bar(['status 1'], [status_1.mean()], color='red')
ax1.bar(['status 2'], [status_2.mean()], color='blue')
ax1.set_xlabel('Survival Status')
ax1.set_ylabel('Average Number of Nodes')

# Plot the first pie chart
ax2.pie(nodes_counts_1, labels=nodes_counts_1.index)
ax2.set_title('Distribution of number of nodes within status 1')

# Plot the second pie chart
ax3.pie(nodes_counts_2, labels=nodes_counts_2.index)
ax3.set_title('Distribution of number of nodes within status 2')



plt.tight_layout()
plt.show()
```

*Conclusion*

Yes there is a Positive corelation between survival and node count.
Patients with higher number of positive axillary nodes, died within 5 year. 





---



**Nodes vs Age**

```python
# extract the data
nodes = df['nodes']
age = df['age']

# create the figure and axis objects
figure, axis = plt.subplots(figsize=(20,10))

# plot the bar chart
axis.bar(age, nodes, color='red', alpha= (0.2))

# plot the scatter chart on top of the bar chart
axis.scatter(age, nodes)

# set the labels for x and y axis
axis.set_xlabel('Age')
axis.set_ylabel('No of Nodes')

# add a title to the plot
axis.set_title('Relationship between Age and No of Nodes')

# show the plot
plt.show()
```

*Conclusion*

There is no clear corelation between the age and no of positive axllary node count.

Although visually we can see that there are more higher values of node count among the in older populus. But the distribution seems even through out the age.

**Age Vs Status**

```python
fig, axs = plt.subplots(2, 2, figsize=(12, 8))

# Plot 1 - Scatter Plot
plt.sca(axs[0, 0])
plt.scatter(df['age'], df['status'], color='blue', alpha=0.5)
plt.xlabel('Age')
plt.ylabel('Survival Status')
plt.title('Age versus Survival Status')

# Plot 2 - Bar Plot
mean_age_status_1 = df[df['status'] == 1]['age'].mean()
mean_age_status_2 = df[df['status'] == 2]['age'].mean()
index = np.arange(2)
bar_width = 0.35
plt.sca(axs[0, 1])
rects1 = plt.bar(index, [mean_age_status_1, mean_age_status_2], bar_width,)
plt.xlabel('Survival Status')
plt.ylabel('Age')
plt.xticks(index)
#plt.xticklabels(['Status 1', 'Status 2'])
plt.title('Average Age by Survival Status')
plt.grid()

# Plot 3 - Pie Chart
status_1 = df[df['status'] == 1]['age']
age_distribution = status_1.value_counts(normalize=True)
plt.sca(axs[1, 0])
plt.pie(age_distribution, labels=age_distribution.index, shadow=True)
plt.title('Age Distribution in Status 1')

# Plot 4 - Pie Chart
status_2 = df[df['status'] == 2]['age']
age_distribution = status_2.value_counts(normalize=True)
plt.sca(axs[1, 1])
plt.pie(age_distribution, labels=age_distribution.index, shadow=True)
plt.title('Age Distribution in Status 2')

plt.tight_layout()
plt.show()
```

*Conclusion*

No clear connclusion can be made reguarding age being directly corelated in bing a factor in survival.

within the given sample and data size there is only a very small difference in young people being able to survive 5 years or longer vs people who died after 5 years.

Visually in the scatter plot it can be seen that the status 2 is slightly shifted towards higher age group but again due to even distribution and small sample size, no solid conclusion can be made.
